# Raport Techniczny: Detekcja Dziur w Nawierzchni (Pothole Detection)

Metoda: Klasyczne Przetwarzanie Obrazu (Computer Vision) bez użycia Sieci Neuronowych.
Podejście oparte na analizie morfologicznej i geometrycznej

### Trudne scenariusze:

- Mokra nawierzchnia: silne refleksy (odbicia nieba)
- Wysoki szum (tekstura): Przez strukture asfaltu algorytmy detekcji krawędzi mogą interpretować małe nierówności jako istotne obiekty
- Zróżnicowane oświetlenie: Rózne obiekty i zmienne warunki pogodowe.

### Pipeline 

- Preprocessing (Szarość i Resize)
- Filtracja Szumu (Gaussian Blur)
- Detekcja Krawędzi (Dynamic Canny)
- Region of Interest (ROI)
- Morfologia Matematyczna (Zamykanie kształtów)
- Filtracja Konturów (Logika decyzyjna)
